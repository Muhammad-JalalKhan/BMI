```python
import pandas as pd  # For data manipulation
import numpy as np  # For numerical computations
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv("used_car_dataset.csv")
```

```python
# Display first few rows
print(df.head())

# Check data types and missing values
print(df.info())
print(df.isnull().sum())

# Summary statistics
print(df.describe())
```

```python
# Remove rows with missing values
df = df.dropna()


```

```python
df.to_csv("processed_dataset.csv", index=False)
from google.colab import files
files.download("processed_dataset.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# First few rows
print(df.head())

# Shape of the dataset
print(f"Dataset has {df.shape[0]} rows and {df.shape[1]} columns.")

# Summary of the dataset
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Statistical summary
print(df.describe())
```

```python
# Count duplicate rows
print(f"Number of duplicate rows: {df.duplicated().sum()}")

# Remove duplicates if necessary
df = df.drop_duplicates()

```

```python
# Distribution plots
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()

```

```python
# Bar plots for categorical data
for col in df.select_dtypes(include=['object']).columns:
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f"Count of {col}")
    plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
# Calculate the correlation matrix only for numeric columns
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Pairplot (for small datasets)
sns.pairplot(df)
plt.show()
```

```python
# Boxplot to see distribution by category
for col in df.select_dtypes(include=['object']).columns:
    for num_col in df.select_dtypes(include=['float64', 'int64']).columns:
        sns.boxplot(x=col, y=num_col, data=df)
        plt.title(f"{num_col} by {col}")
        plt.show()
```

```python
import missingno as msno

# Missing data heatmap
msno.heatmap(df)
plt.show()
```

```python
# Scatter plots for numerical variables
sns.scatterplot(x='feature1', y='feature2', data=df)
plt.title("Feature1 vs Feature2")
plt.show()
```

```python
import matplotlib.pyplot as plt


```

```python
# Histogram
plt.hist(df['Year'], bins=20, color='violet', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar plot
df['AdditionInfo'].value_counts().plot(kind='bar', color='orange')
plt.title("Bar Plot")
plt.xlabel("AdditionInfo")
plt.ylabel("Count")
plt.show()
```

```python
# Boxplot to detect outliers
sns.boxplot(data=df, x='FuelType', y='Year')
plt.title("Boxplot")
plt.show()
```

```python
# Pairplot for exploring relationships
sns.pairplot(df, hue='FuelType')  # Color by a categorical variable
plt.show()
```

```python
# Count plot for categorical data
sns.countplot(x='Transmission', data=df, palette="Set2")
plt.title("Count Plot")
plt.show()
```

```python
# Multiple subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# First plot
sns.histplot(df['Year'], kde=True, ax=axes[0])
axes[0].set_title("Histogram with KDE")

# Second plot
sns.boxplot(data=df, y='Age', ax=axes[1])
axes[1].set_title("Boxplot")

plt.tight_layout()
plt.show()
```

```python
# Violin plot to see distributions
sns.violinplot(x='FuelType', y='Year', data=df, palette="muted")
plt.title("Violin Plot")
plt.show()
```
